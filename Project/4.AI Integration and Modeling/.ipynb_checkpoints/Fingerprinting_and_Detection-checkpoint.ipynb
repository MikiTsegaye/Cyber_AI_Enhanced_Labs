```python
import torch
from sentence_transformers import SentenceTransformer
import pandas as pd

# 1. THE BIG TEST
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"--- Device: {device.upper()} ---")

# 2. LOAD MODEL (Onto GPU if available)
model = SentenceTransformer('all-MiniLM-L6-v2', device=device)

# 3. DEFINE PATH
path = "C:/Users/noaga/OneDrive/Desktop/GitProjects/Cyber_AI_Enhanced_Labs/Project/2.Data preparing and cleaning/Splited_data"

# 4. ENCODE SAMPLES (Check if it works)
df_train = pd.read_csv(f"{path}/01_train_normal.csv")
print(f"Dataset loaded. Total rows: {len(df_train)}")

# Encode a small batch to test speed
X_train = model.encode(df_train['clean_text'].head(1000).fillna("").tolist(), 
                       show_progress_bar=True, 
                       convert_to_tensor=True)

print(f"Test Successful. Tensor Shape: {X_train.shape}")
```

```python
def encode_and_save(csv_name):
    df = pd.read_csv(f"{path}/{csv_name}")
    sentences = df['clean_text'].fillna("").astype(str).tolist()
    print(f"Encoding {csv_name} ({len(sentences)} rows)...")
    
    # batch_size=128 is a "sweet spot" for the RTX 3050
    embeddings = model.encode(sentences, 
                              batch_size=128, 
                              show_progress_bar=True, 
                              convert_to_tensor=True).cpu() # Move to RAM immediately
    return embeddings

# 1. Fingerprinting Data: 100% Normal emails
X_train_emb = encode_and_save("01_train_normal.csv")

# 2. Mixed Detection Data: Normal + Phishing
X_test_emb = encode_and_save("03_test_all.csv")
y_test = pd.read_csv(f"{path}/03_test_all.csv")['is_phishing'].values
```

```python
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# 1. THE ARCHITECTURE (First version - Deep AE)
class SpearPhishingAE(nn.Module):
    def __init__(self, input_dim=384):
        super(SpearPhishingAE, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 32) # The Fingerprint
        )
        self.decoder = nn.Sequential(
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.Linear(64, 128),
            nn.ReLU(),
            nn.Linear(128, input_dim) # Reconstruction
        )

    def forward(self, x):
        return self.decoder(self.encoder(x))

# 2. INITIALIZATION
model_ae = SpearPhishingAE().to(device)
criterion = nn.MSELoss() # Required for measuring anomaly deviation
optimizer = optim.Adam(model_ae.parameters(), lr=1e-3)

# 3. TRAINING DATA SETUP
train_loader = DataLoader(TensorDataset(X_train_emb), batch_size=512, shuffle=True)

# 4. THE TRAINING LOOP
print(f"Starting Fingerprinting on {device}...")
model_ae.train()
for epoch in range(20):
    total_loss = 0
    for batch in train_loader:
        inputs = batch[0].to(device)
        optimizer.zero_grad()
        outputs = model_ae(inputs)
        loss = criterion(outputs, inputs) # MSE Error
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}/20 | Reconstruction Loss: {total_loss/len(train_loader):.6f}")
```

```python
import numpy as np

# 1. Evaluate the model on the TEST set (Mixed data)
model_ae.eval()
with torch.no_grad():
    # Pass the test embeddings through the brain
    reconstructed = model_ae(X_test_emb.to(device))
    # Calculate Mean Squared Error for every single email individually
    mse_errors = torch.mean((X_test_emb.to(device) - reconstructed)**2, dim=1).cpu().numpy()

print(f"Computed anomaly scores for {len(mse_errors)} emails.")

# 2. Threshold Tuning (The Reward Stage)
# We'll start by looking at the 95th percentile (flagging the top 5% most weird emails)
threshold = np.percentile(mse_errors, 95)
y_pred = [1 if e > threshold else 0 for e in mse_errors]

print(f"Current Threshold set at: {threshold:.6f}")
```

```python
!pip -q install seaborn matplotlib
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Generate Metrics
report = classification_report(y_test, y_pred, target_names=['Normal', 'Phishing'])
conf_matrix = confusion_matrix(y_test, y_pred)

print("--- Group 7 Detection Performance ---")
print(report)

# 2. Visualize the Confusion Matrix
plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Predicted Normal', 'Predicted Phishing'],
            yticklabels=['Actual Normal', 'Actual Phishing'])
plt.title('Spear Phishing Detection: Confusion Matrix')
plt.show()
```

```python
# Try making the model more sensitive by flagging the top 20% or 30% of weirdest emails
# Adjust this number (e.g., 80, 75, 70) to see the Recall jump up!
tuned_percentile = 70 
new_threshold = np.percentile(mse_errors, tuned_percentile)
y_pred_tuned = [1 if e > new_threshold else 0 for e in mse_errors]

print(f"--- Results after Tuning to {100-tuned_percentile}% Sensitivity ---")
print(classification_report(y_test, y_pred_tuned, target_names=['Normal', 'Phishing']))
```

```python
# Create a temporary dataframe to link errors back to text
df_test = pd.read_csv(f"{path}/03_test_all.csv")
df_test['anomaly_score'] = mse_errors

# Get the top 5 highest error emails (The most anomalous)
top_anomalies = df_test.sort_values(by='anomaly_score', ascending=False).head(5)

print("--- xAI: Top 5 Anomalous Emails Detected ---")
for i, row in top_anomalies.iterrows():
    print(f"\nScore: {row['anomaly_score']:.6f}")
    print(f"Text snippet: {row['clean_text'][:200]}...")
    print("-" * 30)
```

```python
import pandas as pd

def get_metrics(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tp = cm[1, 1]
    fn = cm[1, 0]
    fp = cm[0, 1]
    tn = cm[0, 0]
    recall = tp / (tp + fn)
    fpr = fp / (fp + tn)
    return recall, fpr, fn

# Metrics for your original 95th percentile
rec1, fpr1, fn1 = get_metrics(y_test, y_pred)

# Metrics for the Tuned (70th percentile) version
rec2, fpr2, fn2 = get_metrics(y_test, y_pred_tuned)

results_table = pd.DataFrame({
    "Metric": ["Recall (Detection Rate)", "False Positive Rate (FPR)", "False Negatives (Missed)"],
    "Initial (Strict)": [f"{rec1:.2%}", f"{fpr1:.2%}", fn1],
    "Tuned (Morpheus Optimized)": [f"{rec2:.2%}", f"{fpr2:.2%}", fn2]
})

print("--- Final Comparison for Group 7 Report ---")
print(results_table.to_string(index=False))
```
